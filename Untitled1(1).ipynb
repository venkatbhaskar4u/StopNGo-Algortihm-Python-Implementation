```python
def sender():
    global d,time,frameCounter
    print(f"Sender Window: [{d['alow']},{d['aHigh']}]")
    if (frames[time-7]==0 & time-7>0):
        print("frame:",time-7)
    else:
        if frames[time]:
            print(f"Sender Window: [{d['alow']},{d['aHigh']}], transmitting frame {frameCounter}, good transmission")
            frameCounter+=1
        else:
            print(f"Sender Window: [{d['alow']},{d['aHigh']}], transmitting frame {frameCounter}, bad transmission")
            frameCounter+=1
        if frameCounter>d['aHigh']:
            frameCounter=d['alow']
```

```python
def reciever():
    global time,d,frameCounter2
    if time>3:
        if frames[time-4]==1:
            print(f"Reciever got frame: {frameCounter2}, Sending ACK {d['alow']}, good transmission")
            if d['retransmit']==0:
                d['alow']+=1
                d['aHigh']+=1
        else:
            d['retransmit']=1
        frameCounter2+=1
    if frameCounter2>d['aHigh']:
        frameCounter2=d['alow']
```

```python
while time<30:
    print("time",time)
    sender()
    reciever()
    time+=1
```
